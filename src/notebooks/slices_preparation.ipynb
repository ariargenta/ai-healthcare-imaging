# Create groups of slices

```python
from glob import glob
import shutil
import os
import dicom2nifti
import nibabel as nib
import numpy as np
```

```python
in_path = '../../datasets/DICOM/raw_files/slices'
out_path = '../../datasets/DICOM/dataset_groups/slices'
```

```python
for patient in glob(in_path + '/*'):
    patient_name = os.path.basename(os.path.normpath(patient))
    number_folders = int(len(glob(patient + '/*')) / 64)

    for i in range(number_folders):
        output_path_name = os.path.join(out_path, patient_name + '_' + str(i))
        os.mkdir(output_path_name)
        
        for i, file in enumerate(glob(patient + '/*')):
            if i == 64 + 1:
                break
            shutil.move(file, output_path_name)
```

# Convert DICOM files into NIfTI

```python
in_slices_path = '../../datasets/DICOM/dataset_groups/slices/*'
in_labels_path = '../../datasets/DICOM/dataset_groups/labels/*'
out_slices_path  ='../../datasets/NIfTI/slices'
out_labels_path = '../../datasets/NIfTI/labels'

list_images = glob(in_slices_path)
list_labels = glob(in_labels_path)
```

```python
for patient in list_labels:
    patient_name = os.path.basename(os.path.normpath(patient))
    dicom2nifti.dicom_series_to_nifti(patient, os.path.join(out_labels_path, patient_name + '.nii.gz'))
```
